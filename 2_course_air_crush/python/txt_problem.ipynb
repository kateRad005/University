```python
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
df = pd.read_csv("newCrash.csv")
```

```python
df
```

```python
nums = ["1","2","3","4","5","6","7","8","9","0"]
```

```python
modal_verbs = ["can", "could", "need", "may", "able", "must", "have", "should", "ought","shall", "will","would", "used","to"]
```

```python
punctuation = [".", ",","'", "!", "?",")","("]
```

```python
def stop_words(w:str):
    if w in modal_verbs:
        return False
    elif w in stopwords.words('english'):
        return False
    elif w in punctuation:
        return False
    for elem in nums:
        if elem in w:
            return False
    else:
        return True
```

```python
dct = {}
ps = PorterStemmer()
for elem in df["Summary"]:
    if  type(elem)!= float:
        elem = elem.replace(",", "")
        sentences = sent_tokenize(elem)
        for sent in sentences:
            sent = sent.replace("!", "")
            sent = sent.replace("?", "")
            for word in word_tokenize(sent):
                w = word.lower()
                if stop_words(w):
                    if w in dct.keys():
                        dct[w] += 1
                    else:
                        dct[w] = 1
```

```python
sort_dct = sorted(dct.items(), key=lambda x: x[1], reverse=True)
```

```python
len(sort_dct)
```

```python
words = {"Word":list(dct.keys()),"Count":list(dct.values())}
```

```python
words_df = pd.DataFrame(words)
```
